Generate white noise and get classification images. white_noise() function takes in a model and the data with generation parameters to create classification images. Gamma = 0 produces white noise, and gamma = 1 returns the mean data per classification from training.

```python
from noise_gen import white_noise
from noise_gen import plot_avg_noise

# Change this value to adjust the gamma for all models when testing
gamma = 0.5

# noise, stats, predictions, targets = whitenoise(model, data_loader, mean_data, gamma, iterations per class, batch_size)
ctrl_MNIST_noise, ctrl_MNIST_stats, ctrl_MNIST_pred, ctrl_MNIST_targets = white_noise(ctrl_model, MNIST_train_loader, MNIST_mean_imgs, gamma, 10, 100)
mdl1_MNIST_noise, mdl1_MNIST_stats, mdl1_MNIST_pred, mdl1_MNIST_targets = white_noise(model_1, MNIST_train_loader, MNIST_mean_imgs, gamma, 10, 100)
mdl2_MNIST_noise, mdl2_MNIST_stats, mdl2_MNIST_pred, mdl2_MNIST_targets = white_noise(model_2, MNIST_train_loader, MNIST_mean_imgs, gamma, 10, 100)

# plot and return average noise tensors per class
ctrl_MNIST_mean_noise = plot_avg_noise(ctrl_MNIST_noise, MNIST_train_loader)
mdl1_MNIST_mean_noise = plot_avg_noise(mdl1_MNIST_noise, MNIST_train_loader)
mdl2_MNIST_mean_noise = plot_avg_noise(mdl2_MNIST_noise, MNIST_train_loader)

# tensorise predictions and targets for Confusion Matrix calculations
ctrl_MNIST_pred = torch.cat(ctrl_MNIST_pred)
ctrl_MNIST_targets = torch.Tensor(ctrl_MNIST_targets)
mdl1_MNIST_pred = torch.cat(mdl1_MNIST_pred)
mdl1_MNIST_targets = torch.Tensor(mdl1_MNIST_targets)
mdl2_MNIST_pred = torch.cat(mdl2_MNIST_pred)
mdl2_MNIST_targets = torch.Tensor(mdl2_MNIST_targets)
```

# Kernel Weights on MNIST Data
The next parts have been computed manually in the notebook, as different models have different number of convolutional layers, meaning they have different numbers of activations. This is difficult to account for in a generalized function, as the models output a different number of layers, making the code more error prone if not accounted for manually (lack of clarity). I have considered using a return tuple with indexing, but without names and labels, it gets confusing fast. For the sake of clarity, I have individually accounted for each layer, but for production code or for mass distribution and use, it is possible to write these next few cells into functions and to call them individually and index their returns to work with networks of variable layers. Apologies for the inconvenience, just know that the code ahead repeats quite a bit.

```python
from tqdm import tqdm
from noise_gen import padded_rf # This helps get receptive features for kernel visualization

ctrl_model.eval()
ctrl_act1_list = []
ctrl_act2_list = []

# Run noise data back through the control model
batch_size = 100
data = torch.cat(list(ctrl_MNIST_noise.values()))
iter_ = len(data)//batch_size
for i in tqdm(range(iter_), desc='Evaluating...'):
    input = data[i*batch_size:(i+1)*batch_size]
    if cuda:
        input = input.type_as(torch.FloatTensor()).cuda()
    else:
        input = input.type_as(torch.FloatTensor())
    with torch.no_grad():
        ctrl_out,ctrl_act2,ctrl_act1 = ctrl_model(input[:,None,...])
    ctrl_act1_list.append(ctrl_act1[:,:,5,5].cpu())
    ctrl_act2_list.append(ctrl_act2[:,:,5,5].cpu())

# Track the activation maps from the noise in convolutional layers
ctrl_conv1 = torch.cat(ctrl_act1_list)
ctrl_conv2 = torch.cat(ctrl_act2_list)

# Format data
rf_info = {'conv1':[2.5, 1, 5, 24, 20], 'conv2':[7, 2, 14, 8, 50]}
img_num = len(ctrl_MNIST_noise.values())
conv1_rf = torch.zeros(ctrl_conv1.shape[1], 5, 5)
conv2_rf = torch.zeros(ctrl_conv2.shape[1], 14, 14)
for i in tqdm(range(img_num), desc='Formatting...'):
    conv1_rf += padded_rf(data[i].type_as(torch.FloatTensor()), 'conv1', 5, 5, rf_info)[None, ...].repeat(ctrl_conv1.shape[1], 1, 1) * ctrl_conv1[i][..., None, None]
    conv2_rf += padded_rf(data[i].type_as(torch.FloatTensor()), 'conv2', 5, 5, rf_info)[None, ...].repeat(ctrl_conv2.shape[1], 1, 1) * ctrl_conv2[i][..., None, None]

# Visualize kernel on noise
plt.suptitle('Control Model Activations of Convolution Layer 1 on MNIST Noise')
for i in range(conv1_rf.shape[0]):
    plt.subplot(4, 5, i+1)
    plt.axis('off')
    plt.imshow(conv1_rf[i].numpy()/(img_num))
plt.show()

plt.suptitle('Control Model Activations of Convolution Layer 2 on MNIST Noise')
for i in range(conv2_rf.shape[0]):
    plt.subplot(5, 10, i+1)
    plt.axis('off')
    plt.imshow(conv2_rf[i].numpy()/(img_num))
plt.show()
```

```python
# Generate noise data 5 times for averaging
for i in range(file_num):
    gen_ctrl_noise(ctrl_model, iters, all_size, batch_size, p, num_cls, i, 'ctrl_MNIST_noise_acts')
```

```python
# Initialize data types to store activation maps
conv2_act_noise = {}
conv1_act_noise = {}
noise = {}

for i in range(num_cls):
    conv2_act_noise[i] = []
    conv1_act_noise[i] = []
    noise[i] = []

# Extract noise activations from file
for i in range(1, file_num+1):
    with open('cache/ctrl_MNIST_noise_acts{}.pkl'.format(i), 'rb') as f:
        cur_data = pickle.load(f)
    conv2_act = cur_data['conv2']
    conv1_act = cur_data['conv1']
    noise_img = cur_data['img']
    for k in range(num_cls):
        conv2_act_noise[k].append(conv2_act[k][None, ...])
        conv1_act_noise[k].append(conv1_act[k][None, ...])
        noise[k].append(noise_img[k][None, ...])

# Reformat extracted data
for i in range(num_cls):
    conv2_act_noise[i] = torch.cat(conv2_act_noise[i]).mean(0)
    conv1_act_noise[i] = torch.cat(conv1_act_noise[i]).mean(0)
    noise[i] = torch.cat(noise[i]).mean(0)
```

```python
# Initialize datatypes for storing activations
pred_noise_avg = []
conv2_act_noise_avg = {}
conv1_act_noise_avg = {}
ctrl_model.eval()

# activation of white noise average
for i in range(num_cls):
    with torch.no_grad():
        cur_avg_noise = noise[i][None, ...]
        if cuda:
            cur_avg_noise = cur_avg_noise.cuda()
        out_noise_avg, conv2_act_noise_avg[i], conv1_act_noise_avg[i] = ctrl_model(cur_avg_noise)
        
    pred_noise_avg.append(out_noise_avg.max(1)[1].cpu().numpy()[0])
    conv2_act_noise_avg[i] = conv2_act_noise_avg[i][0].cpu()
    conv1_act_noise_avg[i] = conv1_act_noise_avg[i][0].cpu()
    
pred_noise_avg
```

# Adversarial Defense - Task 5

Here we make a small perturbation to a specific class as a predictable control form of an adversarial attack using a patch, or a small form of white noise as a given pattern. We will try to apply a classification image transform, overlayed on top of the original input to try and sway the model into making the correct classification. Other possible forms of preventing an adversarial attack include gradient masking or training the model with preprocessed data with white noise included in the training data. Another method would be the use of ensemble training, however, there are computational limitations. Given the nature of this project and the previous tasks, it appears appropriate to introduce noise through the classification images as the form of adversarial defense.
